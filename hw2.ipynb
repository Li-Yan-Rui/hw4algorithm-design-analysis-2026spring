## A 排序

```python
## add your code here
import sys

sys.setrecursionlimit(2000)

def solve():
    input_data = sys.stdin.read().split()
    if not input_data:
        return

    n = int(input_data[0])
    a = int(input_data[1])
    b = int(input_data[2])

    p = [0] * n
    for i in range(n):
        p[i] = int(input_data[3 + i])

    ops = []

    def swp():
        nonlocal p
        ops.append(0)
        p = [b if v == a else (a if v == b else v) for v in p]

    def add(x):
        nonlocal p
        if x == 0:
            return
        ops.append(x)
        p = [(v + x) % n for v in p]

    def xxr(x):
        nonlocal p
        if x == 0:
            return
        ops.append(-x)
        p = [v ^ x for v in p]

    def getPs(aa, bb):
        d = (bb - aa + n - L + n) % n
        pa = 0
        pb = 0
        s = n // 2
        while s >= 2 * L:
            if d >= s:
                d -= s
                pb += s // 2
            else:
                pa += s // 2
            s //= 2

        pa += n // 2 + (aa & (L - 1))
        pb += (aa & (L - 1))
        return pa, pb

    def swapEl(c, d):
        if (c // L) % 2 == (d // L) % 2:
            if (c // L) % 2 == 0:
                t = (c & (L - 1)) + L
            else:
                t = (c & (L - 1))

            swapEl(c, t)
            swapEl(d, t)
            swapEl(c, t)
        else:
            pa, _ = getPs(a, b)
            pc, _ = getPs(c, d)

            add((pc - c + n) % n)
            xxr(pc ^ pa)
            add((a - pa + n) % n)
            swp()
            add((pa - a + n) % n)
            xxr(pc ^ pa)
            add((c - pc + n) % n)

    class P:
        __slots__ = ['arr', 'sz', 'v']

        def __init__(self):
            self.arr = [0] * 1024
            self.sz = 0
            self.v = []

        def build(self):
            if self.sz == 1:
                return True

            ev = P()
            od = P()
            half = self.sz // 2

            for i in range(half):
                ev.arr[i] = self.arr[2 * i] // 2
                od.arr[i] = self.arr[2 * i + 1] // 2

            ev.sz = half
            od.sz = half

            if not ev.build() or not od.build():
                return False

            if (self.arr[0] & 1) != 0:
                self.v.append(1 if self.sz == 2 else -1)

            te = 0
            to = 0

            for x in ev.v:
                if x > 0:
                    self.v.append(-1)
                    self.v.append(1)
                else:
                    self.v.append(2 * x)
                    te ^= -2 * x

            if te != 0:
                self.v.append(-te)

            for x in od.v:
                if x > 0:
                    self.v.append(1)
                    self.v.append(-1)
                else:
                    self.v.append(2 * x)
                    to ^= -2 * x

            if (to & half) != (te & half):
                for _ in range(self.sz // 4):
                    self.v.append(-1)
                    self.v.append(1)

            if te >= half:
                te -= half
            if to >= half:
                to -= half

            if te != to:
                return False

            tmp = []
            for x in self.v:
                if not tmp:
                    tmp.append(x)
                elif x < 0 and tmp[-1] < 0:
                    last = tmp[-1]
                    merged = -((-last) ^ (-x))

                    if merged == 0:
                        tmp.pop()
                    else:
                        tmp[-1] = merged
                else:
                    tmp.append(x)

            self.v = tmp
            return True

    L = (a - b + n) % n
    L = L & -L

    if L == 0:
        L = n

    if L > 1:
        sub = P()
        sub.sz = L

        for i in range(n):
            sub.arr[i] = p[i] & (L - 1)

        if not sub.build():
            sys.stdout.write("-1\n")
            return

        for x in sub.v:
            if x > 0:
                add(x)
            else:
                xxr(-x)

    for r in range(L):
        tmp = [p[j] for j in range(r, n, L)]
        tmp.sort()

        idx = 0
        for j in range(r, n, L):
            if tmp[idx] != j:
                sys.stdout.write("-1\n")
                return
            idx += 1

        for j in range(r, n, L):
            if p[j] != j:
                swapEl(j, p[j])

    for i in range(n):
        if p[i] != i:
            raise AssertionError()

    out = [str(len(ops))]
    for x in ops:
        if x == 0:
            out.append("0")
        elif x < 0:
            out.append(f"1 {-x}")
        else:
            out.append(f"2 {x}")

    sys.stdout.write('\n'.join(out) + '\n')

if __name__ == '__main__':
    solve()
```

## B 长跑

```python
## add your code here
#include <bits/stdc++.h>
using namespace std;

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int N;
    long long L, Maxn, S;

    while (cin >> N >> L >> Maxn >> S) {
        unordered_map<long long, long long> station_map;
        station_map.reserve(N * 2);

        for (int i = 0; i < N; ++i) {
            long long p, c;
            cin >> p >> c;

            // 只保留起点和终点之间的补给站
            if (p > 0 && p < L) {
                auto it = station_map.find(p);
                if (it == station_map.end() || c < it->second) {
                    station_map[p] = c;
                }
            }
        }

        // 初始体力可以直接到终点
        if (L <= Maxn) {
            cout << "Yes\n";
            continue;
        }

        vector<pair<long long, long long>> stations;
        stations.reserve(station_map.size());

        for (auto &kv : station_map) {
            stations.push_back(kv);
        }

        sort(stations.begin(), stations.end());

        // 单调队列，存储 {位置, 到达该位置的最小花费}
        deque<pair<long long, long long>> q;
        q.push_back({0, 0});

        long long min_cost_to_destination = LLONG_MAX;

        for (auto &[p, c] : stations) {
            // 移除当前补给站无法从其到达的历史状态
            while (!q.empty() && p - q.front().first > Maxn) {
                q.pop_front();
            }

            if (q.empty()) {
                continue;
            }

            // 队首是当前可达范围内花费最小的状态
            long long current_dp = q.front().second + c;

            // 从当前补给站可以直接到终点
            if (L - p <= Maxn) {
                min_cost_to_destination = min(min_cost_to_destination, current_dp);
            }

            // 维护单调队列：花费更高或相同的位置没有保留价值
            while (!q.empty() && q.back().second >= current_dp) {
                q.pop_back();
            }

            q.push_back({p, current_dp});
        }

        if (min_cost_to_destination <= S) {
            cout << "Yes\n";
        } else {
            cout << "No\n";
        }
    }

    return 0;
}
```

## C 最长回文

```python
## add your code here
#include <stdio.h>
#include <stdlib.h>
#include <string.h>

#define BUF_SIZE (1 << 16)

static unsigned char buffer[BUF_SIZE];
static int ptr = 0, len_buf = 0;

static int read_char_fast(void) {
    if (ptr >= len_buf) {
        len_buf = (int)fread(buffer, 1, BUF_SIZE, stdin);
        ptr = 0;
        if (len_buf <= 0) return -1;
    }
    return buffer[ptr++];
}

static int next_int(void) {
    int c;

    do {
        c = read_char_fast();
    } while (c <= ' ' && c != -1);

    int sign = 1;

    if (c == '-') {
        sign = -1;
        c = read_char_fast();
    }

    int val = 0;

    while (c > ' ') {
        val = val * 10 + c - '0';
        c = read_char_fast();
    }

    return val * sign;
}

static void next_string(char *s) {
    int c;

    do {
        c = read_char_fast();
    } while (c <= ' ' && c != -1);

    int idx = 0;

    while (c > ' ') {
        s[idx++] = (char)c;
        c = read_char_fast();
    }

    s[idx] = '\0';
}

static int min_int(int a, int b) {
    return a < b ? a : b;
}

static int max_int(int a, int b) {
    return a > b ? a : b;
}

typedef struct {
    int *odd;
    int *even;
} ManacherResult;

typedef struct {
    int *sa;
    int *rank;
    int *height;
    int *st;
    int *log2val;
    int len;
    int K;
} LCPQuery;

static ManacherResult manacher(const char *s, int n) {
    ManacherResult res;

    res.odd = (int *)calloc(n, sizeof(int));
    res.even = (int *)calloc(n, sizeof(int));

    int l = 0, r = -1;

    for (int i = 0; i < n; i++) {
        int k = i > r ? 1 : min_int(res.odd[l + r - i], r - i + 1);

        while (i - k >= 0 && i + k < n && s[i - k] == s[i + k]) {
            k++;
        }

        res.odd[i] = k;

        if (i + k - 1 > r) {
            l = i - k + 1;
            r = i + k - 1;
        }
    }

    l = 0;
    r = -1;

    for (int i = 0; i < n; i++) {
        int k = i > r ? 0 : min_int(res.even[l + r - i + 1], r - i + 1);

        while (i - k - 1 >= 0 && i + k < n &&
               s[i - k - 1] == s[i + k]) {
            k++;
        }

        res.even[i] = k;

        if (i + k - 1 > r) {
            l = i - k;
            r = i + k - 1;
        }
    }

    return res;
}

static int *build_sa(const int *s, int n) {
    int *sa = (int *)malloc(n * sizeof(int));
    int *x = (int *)malloc(n * sizeof(int));
    int *y = (int *)malloc(n * sizeof(int));

    int max_value = max_int(256, n + 5);
    int *c = (int *)calloc(max_value, sizeof(int));

    int m = 27;

    for (int i = 0; i < n; i++) {
        x[i] = s[i];
        c[x[i]]++;
    }

    for (int i = 1; i < m; i++) {
        c[i] += c[i - 1];
    }

    for (int i = n - 1; i >= 0; i--) {
        sa[--c[x[i]]] = i;
    }

    for (int k = 1; k < n; k <<= 1) {
        int p = 0;

        for (int i = n - k; i < n; i++) {
            y[p++] = i;
        }

        for (int i = 0; i < n; i++) {
            if (sa[i] >= k) {
                y[p++] = sa[i] - k;
            }
        }

        memset(c, 0, m * sizeof(int));

        for (int i = 0; i < n; i++) {
            c[x[y[i]]]++;
        }

        for (int i = 1; i < m; i++) {
            c[i] += c[i - 1];
        }

        for (int i = n - 1; i >= 0; i--) {
            sa[--c[x[y[i]]]] = y[i];
        }

        int *tmp = x;
        x = y;
        y = tmp;

        x[sa[0]] = 0;
        p = 1;

        for (int i = 1; i < n; i++) {
            int a = sa[i];
            int b = sa[i - 1];

            int a1 = y[a];
            int b1 = y[b];

            int a2 = a + k < n ? y[a + k] : -1;
            int b2 = b + k < n ? y[b + k] : -1;

            if (a1 == b1 && a2 == b2) {
                x[a] = p - 1;
            } else {
                x[a] = p++;
            }
        }

        if (p >= n) {
            break;
        }

        m = p;
    }

    free(x);
    free(y);
    free(c);

    return sa;
}

static int *build_height(const int *s, const int *sa, const int *rank, int n) {
    int *height = (int *)calloc(n, sizeof(int));

    int h = 0;

    for (int i = 0; i < n; i++) {
        int r = rank[i];

        if (r == 0) {
            h = 0;
            continue;
        }

        int j = sa[r - 1];

        while (i + h < n && j + h < n && s[i + h] == s[j + h]) {
            h++;
        }

        height[r] = h;

        if (h > 0) {
            h--;
        }
    }

    return height;
}

static void build_sparse_table(LCPQuery *q) {
    int len = q->len;

    q->log2val = (int *)calloc(len + 1, sizeof(int));

    for (int i = 2; i <= len; i++) {
        q->log2val[i] = q->log2val[i >> 1] + 1;
    }

    q->K = q->log2val[len] + 1;
    q->st = (int *)calloc(q->K * len, sizeof(int));

    for (int i = 0; i < len; i++) {
        q->st[i] = q->height[i];
    }

    for (int k = 1; k < q->K; k++) {
        int half = 1 << (k - 1);
        int full = 1 << k;

        for (int i = 0; i + full <= len; i++) {
            int left = q->st[(k - 1) * len + i];
            int right = q->st[(k - 1) * len + i + half];

            q->st[k * len + i] = min_int(left, right);
        }
    }
}

static LCPQuery create_lcp_query(const int *s, int len) {
    LCPQuery q;

    q.len = len;
    q.K = 0;

    q.sa = build_sa(s, len);
    q.rank = (int *)malloc(len * sizeof(int));

    for (int i = 0; i < len; i++) {
        q.rank[q.sa[i]] = i;
    }

    q.height = build_height(s, q.sa, q.rank, len);

    q.st = NULL;
    q.log2val = NULL;

    build_sparse_table(&q);

    return q;
}

static int query_min(LCPQuery *q, int l, int r) {
    int len = q->len;
    int k = q->log2val[r - l + 1];

    int left = q->st[k * len + l];
    int right = q->st[k * len + r - (1 << k) + 1];

    return min_int(left, right);
}

static int lcp(LCPQuery *q, int i, int j) {
    if (i == j) {
        return q->len - i;
    }

    int ri = q->rank[i];
    int rj = q->rank[j];

    if (ri > rj) {
        int tmp = ri;
        ri = rj;
        rj = tmp;
    }

    return query_min(q, ri + 1, rj);
}

static void free_lcp_query(LCPQuery *q) {
    free(q->sa);
    free(q->rank);
    free(q->height);
    free(q->st);
    free(q->log2val);
}

static int lce_back_a_forward_b(LCPQuery *q, int n, int a, int b) {
    if (a < 0 || b < 0 || b >= n) {
        return 0;
    }

    int pos_in_rev_a = n - 1 - a;
    int pos_in_b = n + 1 + b;

    int lcp_val = lcp(q, pos_in_rev_a, pos_in_b);

    int limit = min_int(a + 1, n - b);

    return min_int(lcp_val, limit);
}

int main(void) {
    int n = next_int();

    char *A = (char *)malloc((n + 5) * sizeof(char));
    char *B = (char *)malloc((n + 5) * sizeof(char));

    next_string(A);
    next_string(B);

    int combined_len = 2 * n + 1;
    int *combined = (int *)malloc(combined_len * sizeof(int));

    // reverse(A)
    for (int i = 0; i < n; i++) {
        combined[i] = A[n - 1 - i] - 'A' + 1;
    }

    // separator
    combined[n] = 0;

    // B
    for (int i = 0; i < n; i++) {
        combined[n + 1 + i] = B[i] - 'A' + 1;
    }

    LCPQuery lcp_query = create_lcp_query(combined, combined_len);

    int ans = 0;

    ManacherResult man_a = manacher(A, n);
    int *odd_a = man_a.odd;
    int *even_a = man_a.even;

    ManacherResult man_b = manacher(B, n);
    int *odd_b = man_b.odd;
    int *even_b = man_b.even;

    // 只从 A 中取
    for (int i = 0; i < n; i++) {
        ans = max_int(ans, 2 * odd_a[i] - 1);
    }

    for (int i = 0; i < n; i++) {
        ans = max_int(ans, 2 * even_a[i]);
    }

    // 只从 B 中取
    for (int i = 0; i < n; i++) {
        ans = max_int(ans, 2 * odd_b[i] - 1);
    }

    for (int i = 0; i < n; i++) {
        ans = max_int(ans, 2 * even_b[i]);
    }

    // 中间回文为空，直接 A 向左、B 向右匹配
    for (int k = 0; k < n; k++) {
        int ext = lce_back_a_forward_b(&lcp_query, n, k, k);
        ans = max_int(ans, 2 * ext);
    }

    // 中间回文在 A 中，右端点作为拼接点：奇数回文
    for (int c = 0; c < n; c++) {
        int r = odd_a[c];

        int left = c - r + 1;
        int right = c + r - 1;
        int len = 2 * r - 1;

        int ext = lce_back_a_forward_b(&lcp_query, n, left - 1, right);

        ans = max_int(ans, len + 2 * ext);
    }

    // 中间回文在 A 中，右端点作为拼接点：偶数回文
    for (int c = 0; c < n; c++) {
        int r = even_a[c];

        if (r == 0) continue;

        int left = c - r;
        int right = c + r - 1;
        int len = 2 * r;

        int ext = lce_back_a_forward_b(&lcp_query, n, left - 1, right);

        ans = max_int(ans, len + 2 * ext);
    }

    // 中间回文在 B 中，左端点作为拼接点：奇数回文
    for (int c = 0; c < n; c++) {
        int r = odd_b[c];

        int left = c - r + 1;
        int right = c + r - 1;
        int len = 2 * r - 1;

        int ext = lce_back_a_forward_b(&lcp_query, n, left, right + 1);

        ans = max_int(ans, len + 2 * ext);
    }

    // 中间回文在 B 中，左端点作为拼接点：偶数回文
    for (int c = 0; c < n; c++) {
        int r = even_b[c];

        if (r == 0) continue;

        int left = c - r;
        int right = c + r - 1;
        int len = 2 * r;

        int ext = lce_back_a_forward_b(&lcp_query, n, left, right + 1);

        ans = max_int(ans, len + 2 * ext);
    }

    printf("%d\n", ans);

    free(A);
    free(B);
    free(combined);

    free(odd_a);
    free(even_a);
    free(odd_b);
    free(even_b);

    free_lcp_query(&lcp_query);

    return 0;
}
```

## D 优惠券

```python
## add your code here
#include <stdio.h>
#include <string.h>

#define MAX_X 100005
#define MAX_S 20

int last_pos_arr[MAX_X];
int state_arr[MAX_X];

typedef struct {
    int type; // 0 表示 '?', 1 表示 'I', 2 表示 'O'
    int x;
} Op;

int find_set(int parent[], int i) {
    if (parent[i] == i) return i;
    parent[i] = find_set(parent, parent[i]);
    return parent[i];
}

void solve() {
    int m;

    while (scanf("%d", &m) != EOF) {
        if (m == 0) {
            printf("-1\n");
            continue;
        }

        Op ops[m + 1];
        int parent[m + 2];
        int touched[m + 1];
        int touched_cnt = 0;

        for (int i = 1; i <= m; ++i) {
            char s[MAX_S];
            scanf("%s", s);

            if (s[0] == 'I') {
                int x;
                scanf("%d", &x);

                ops[i].type = 1;
                ops[i].x = x;
                parent[i] = i + 1;
            } else if (s[0] == 'O') {
                int x;
                scanf("%d", &x);

                ops[i].type = 2;
                ops[i].x = x;
                parent[i] = i + 1;
            } else {
                ops[i].type = 0;
                ops[i].x = 0;
                parent[i] = i;
            }
        }

        parent[m + 1] = m + 1;

        int error_line = -1;

        for (int i = 1; i <= m; ++i) {
            if (ops[i].type == 0) continue;

            int type = ops[i].type;
            int x = ops[i].x;

            touched[touched_cnt++] = x;

            if (type == 1) {
                // 购买 I
                if (state_arr[x] == 1) {
                    int p = find_set(parent, last_pos_arr[x] + 1);

                    if (p < i) {
                        parent[p] = p + 1;
                        last_pos_arr[x] = i;
                    } else {
                        error_line = i;
                        break;
                    }
                } else {
                    state_arr[x] = 1;
                    last_pos_arr[x] = i;
                }
            } else if (type == 2) {
                // 使用 O
                if (state_arr[x] == 0) {
                    int p = find_set(parent, last_pos_arr[x] + 1);

                    if (p < i) {
                        parent[p] = p + 1;
                        last_pos_arr[x] = i;
                    } else {
                        error_line = i;
                        break;
                    }
                } else {
                    state_arr[x] = 0;
                    last_pos_arr[x] = i;
                }
            }
        }

        if (error_line != -1) {
            printf("%d\n", error_line);
        } else {
            printf("-1\n");
        }

        for (int i = 0; i < touched_cnt; ++i) {
            int x = touched[i];
            last_pos_arr[x] = 0;
            state_arr[x] = 0;
        }
    }
}

int main() {
    solve();
    return 0;
}
```

## E 任意点

```python
## add your code here
#include <bits/stdc++.h>
using namespace std;

class DSU {
public:
    vector<int> parent, sz;

    DSU(int n) {
        parent.resize(n);
        sz.assign(n, 1);
        iota(parent.begin(), parent.end(), 0);
    }

    int find(int x) {
        if (parent[x] == x) return x;
        return parent[x] = find(parent[x]);
    }

    void unite(int a, int b) {
        int pa = find(a);
        int pb = find(b);
        if (pa == pb) return;

        if (sz[pa] < sz[pb]) swap(pa, pb);
        parent[pb] = pa;
        sz[pa] += sz[pb];
    }
};

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    cin >> n;

    vector<int> x(n), y(n);

    for (int i = 0; i < n; i++) {
        cin >> x[i] >> y[i];
    }

    DSU dsu(n);

    // 横坐标相同或纵坐标相同的点，可以互相到达
    for (int i = 0; i < n; i++) {
        for (int j = i + 1; j < n; j++) {
            if (x[i] == x[j] || y[i] == y[j]) {
                dsu.unite(i, j);
            }
        }
    }

    int components = 0;

    for (int i = 0; i < n; i++) {
        if (dsu.find(i) == i) {
            components++;
        }
    }

    // k 个连通块至少需要加 k - 1 个点连接起来
    cout << components - 1 << '\n';

    return 0;
}
```

## F 通配符匹配

```python
## add your code here
#include <bits/stdc++.h>
using namespace std;

struct Block {
    int id;
    int offset;
};

struct Segment {
    int len;
    vector<Block> blocks;
};

vector<string> blockStrs;
vector<vector<int>> prefixTable;

vector<int> buildPrefix(const string& p) {
    int n = p.size();
    vector<int> nxt(n, 0);

    for (int i = 1; i < n; i++) {
        int j = nxt[i - 1];

        while (j > 0 && p[i] != p[j]) {
            j = nxt[j - 1];
        }

        if (p[i] == p[j]) {
            j++;
        }

        nxt[i] = j;
    }

    return nxt;
}

void kmpSearch(
    const string& text,
    const string& pat,
    const vector<int>& nxt,
    vector<int>& positions,
    vector<unsigned char>& exist
) {
    int n = text.size();
    int m = pat.size();

    if (m > n) return;

    int j = 0;

    for (int i = 0; i < n; i++) {
        while (j > 0 && text[i] != pat[j]) {
            j = nxt[j - 1];
        }

        if (text[i] == pat[j]) {
            j++;
        }

        if (j == m) {
            int pos = i - m + 1;
            positions.push_back(pos);
            exist[pos] = 1;
            j = nxt[j - 1];
        }
    }
}

bool segmentMatchAt(
    const Segment& seg,
    int pos,
    const string& s,
    const vector<vector<unsigned char>>& exist
) {
    int n = s.size();

    if (pos < 0 || pos + seg.len > n) {
        return false;
    }

    for (const auto& b : seg.blocks) {
        int start = pos + b.offset;
        if (!exist[b.id][start]) {
            return false;
        }
    }

    return true;
}

int findSegment(
    const Segment& seg,
    int minPos,
    int limit,
    const string& s,
    const vector<vector<int>>& positions,
    const vector<vector<unsigned char>>& exist
) {
    if (minPos + seg.len > limit) {
        return -1;
    }

    if (seg.blocks.empty()) {
        return minPos;
    }

    int bestBlockIndex = 0;

    for (int i = 1; i < (int)seg.blocks.size(); i++) {
        int id1 = seg.blocks[i].id;
        int id2 = seg.blocks[bestBlockIndex].id;

        if (positions[id1].size() < positions[id2].size()) {
            bestBlockIndex = i;
        }
    }

    int blockId = seg.blocks[bestBlockIndex].id;
    int offset = seg.blocks[bestBlockIndex].offset;

    const vector<int>& posList = positions[blockId];

    auto it = lower_bound(posList.begin(), posList.end(), minPos + offset);

    while (it != posList.end()) {
        int start = *it - offset;

        if (start + seg.len > limit) {
            break;
        }

        if (segmentMatchAt(seg, start, s, exist)) {
            return start;
        }

        ++it;
    }

    return -1;
}

bool matchFile(
    const string& s,
    const vector<Segment>& segments,
    bool startWithStar,
    bool endWithStar,
    const vector<vector<int>>& positions,
    const vector<vector<unsigned char>>& exist
) {
    int n = s.size();
    int k = segments.size();

    if (k == 0) {
        return true;
    }

    int left = 0;
    int right = k - 1;
    int cur = 0;
    int limit = n;

    if (!startWithStar) {
        if (!segmentMatchAt(segments[left], 0, s, exist)) {
            return false;
        }

        cur = segments[left].len;
        left++;
    }

    if (!endWithStar) {
        if (right < left) {
            return cur == n;
        }

        int suffixStart = n - segments[right].len;

        if (suffixStart < 0) {
            return false;
        }

        if (!segmentMatchAt(segments[right], suffixStart, s, exist)) {
            return false;
        }

        limit = suffixStart;
        right--;
    }

    for (int i = left; i <= right; i++) {
        int pos = findSegment(segments[i], cur, limit, s, positions, exist);

        if (pos == -1) {
            return false;
        }

        cur = pos + segments[i].len;
    }

    return cur <= limit;
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    string pattern;
    cin >> pattern;

    int n;
    cin >> n;

    bool startWithStar = pattern.front() == '*';
    bool endWithStar = pattern.back() == '*';

    map<string, int> blockId;
    vector<Segment> segments;

    int m = pattern.size();
    int i = 0;

    while (i < m) {
        if (pattern[i] == '*') {
            i++;
            continue;
        }

        int start = i;

        while (i < m && pattern[i] != '*') {
            i++;
        }

        string part = pattern.substr(start, i - start);

        Segment seg;
        seg.len = part.size();

        int j = 0;

        while (j < (int)part.size()) {
            if (part[j] == '?') {
                j++;
                continue;
            }

            int blockStart = j;

            while (j < (int)part.size() && part[j] != '?') {
                j++;
            }

            string block = part.substr(blockStart, j - blockStart);

            if (!blockId.count(block)) {
                int id = blockStrs.size();
                blockId[block] = id;
                blockStrs.push_back(block);
            }

            seg.blocks.push_back({blockId[block], blockStart});
        }

        segments.push_back(seg);
    }

    int blockCount = blockStrs.size();

    prefixTable.resize(blockCount);

    for (int i = 0; i < blockCount; i++) {
        prefixTable[i] = buildPrefix(blockStrs[i]);
    }

    while (n--) {
        string filename;
        cin >> filename;

        int len = filename.size();

        vector<vector<int>> positions(blockCount);
        vector<vector<unsigned char>> exist(
            blockCount,
            vector<unsigned char>(len + 1, 0)
        );

        for (int i = 0; i < blockCount; i++) {
            kmpSearch(
                filename,
                blockStrs[i],
                prefixTable[i],
                positions[i],
                exist[i]
            );
        }

        if (matchFile(
                filename,
                segments,
                startWithStar,
                endWithStar,
                positions,
                exist
            )) {
            cout << "YES\n";
        } else {
            cout << "NO\n";
        }
    }

    return 0;
}
```

## G 汉诺塔

```python
## add your code here
#include <bits/stdc++.h>
using namespace std;

using int64 = long long;

bool finished(const vector<vector<int>>& peg, int n) {
    if ((int)peg[1].size() == n) return true; // 全部到 B
    if ((int)peg[2].size() == n) return true; // 全部到 C
    return false;
}

int simulateSmall(int n, const vector<string>& pri) {
    vector<vector<int>> peg(3);

    // A 柱初始从大到小放置，back() 表示柱顶
    for (int i = n; i >= 1; i--) {
        peg[0].push_back(i);
    }

    int last = -1;
    int steps = 0;

    while (!finished(peg, n)) {
        for (const string& op : pri) {
            int from = op[0] - 'A';
            int to = op[1] - 'A';

            if (peg[from].empty()) continue;

            int disk = peg[from].back();

            // 不能移动上一次移动过的盘子
            if (disk == last) continue;

            // 目标柱为空，或者目标柱顶盘比当前盘大
            if (peg[to].empty() || peg[to].back() > disk) {
                peg[from].pop_back();
                peg[to].push_back(disk);
                last = disk;
                steps++;
                break;
            }
        }
    }

    return steps;
}

int64 powerInt(int64 base, int exp) {
    int64 ans = 1;
    while (exp--) {
        ans *= base;
    }
    return ans;
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    cin >> n;

    vector<string> pri(6);
    for (int i = 0; i < 6; i++) {
        cin >> pri[i];
    }

    /*
        对任意优先级，步数序列只会出现三种形式：

        1) 1, 3, 7, 15, ...      => 2^n - 1
        2) 1, 3, 9, 27, ...      => 3^(n - 1)
        3) 1, 5, 17, 53, ...     => 2 * 3^(n - 1) - 1

        因此只需要模拟 n = 2 和 n = 3 的小规模情况，即可判断属于哪一种。
    */

    int step2 = simulateSmall(2, pri);
    int step3 = simulateSmall(3, pri);

    int64 ans;

    if (step2 == 5) {
        ans = 2 * powerInt(3, n - 1) - 1;
    } else {
        if (step3 == 7) {
            ans = powerInt(2, n) - 1;
        } else {
            ans = powerInt(3, n - 1);
        }
    }

    cout << ans << '\n';

    return 0;
}
```

## H 马步距离

```python
## add your code here
#include <bits/stdc++.h>
using namespace std;

using ll = long long;

ll knightDistance(ll x1, ll y1, ll x2, ll y2) {
    ll x = llabs(x1 - x2);
    ll y = llabs(y1 - y2);

    // 利用对称性，只考虑 x >= y >= 0
    if (x < y) swap(x, y);

    // 特殊情况：从 (0,0) 到 (1,0)
    if (x == 1 && y == 0) return 3;

    // 特殊情况：从 (0,0) 到 (2,2)
    if (x == 2 && y == 2) return 4;

    ll ans = max((x + 1) / 2, (x + y + 2) / 3);

    if ((ans + x + y) % 2 != 0) {
        ans++;
    }

    return ans;
}

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    ll xp, yp, xs, ys;
    cin >> xp >> yp >> xs >> ys;

    cout << knightDistance(xp, yp, xs, ys) << '\n';

    return 0;
}
```

## I 直方图最大矩形

```python
## add your code here
/**
 * 代码中的类名、方法名、参数名已经指定，请勿修改，直接返回方法规定的值即可
 *
 *
 * @param heights int整型一维数组
 * @param heightsLen int heights数组长度
 * @return int整型
 *
 * C语言声明定义全局变量请加上static，防止重复定义
 */
int largestRectangleArea(int* heights, int heightsLen ) {
    if (heights == 0 || heightsLen == 0) {
        return 0;
    }

    int n = heightsLen;
    int maxArea = 0;

    int stack[heightsLen + 1];
    int top = -1;

    for (int i = 0; i <= n; i++) {
        int curHeight = (i == n) ? 0 : heights[i];

        while (top != -1 && curHeight < heights[stack[top]]) {
            int height = heights[stack[top--]];

            int leftLessIndex = (top == -1) ? -1 : stack[top];
            int width = i - leftLessIndex - 1;

            int area = height * width;
            if (area > maxArea) {
                maxArea = area;
            }
        }

        stack[++top] = i;
    }

    return maxArea;
}
```

## J 消防局的设立

```python
## add your code here
#include <bits/stdc++.h>
using namespace std;

int main() {
    ios::sync_with_stdio(false);
    cin.tie(nullptr);

    int n;
    cin >> n;

    vector<int> parent(n + 1, 0);
    vector<int> depth(n + 1, 0);

    int maxDepth = 0;

    for (int i = 2; i <= n; i++) {
        cin >> parent[i];
        depth[i] = depth[parent[i]] + 1;
        maxDepth = max(maxDepth, depth[i]);
    }

    vector<int> head(maxDepth + 1, -1);
    vector<int> nxt(n + 1, -1);

    for (int i = 1; i <= n; i++) {
        nxt[i] = head[depth[i]];
        head[depth[i]] = i;
    }

    vector<int> hasStation(n + 1, 0);
    vector<int> childStation(n + 1, 0);
    vector<int> grandChildStation(n + 1, 0);

    auto covered = [&](int x) -> bool {
        int p = parent[x];
        int gp = p ? parent[p] : 0;

        // 自己、父亲、祖父有消防局
        if (hasStation[x]) return true;
        if (p && hasStation[p]) return true;
        if (gp && hasStation[gp]) return true;

        // 儿子、孙子有消防局
        if (childStation[x] > 0) return true;
        if (grandChildStation[x] > 0) return true;

        // 兄弟节点有消防局，距离也是 2
        if (p && childStation[p] > 0) return true;

        return false;
    };

    int answer = 0;

    // 从最深的基地开始处理
    for (int d = maxDepth; d >= 0; d--) {
        for (int x = head[d]; x != -1; x = nxt[x]) {
            if (covered(x)) continue;

            // 如果 x 没被覆盖，最优策略是在 x 的祖父处建消防局
            // 如果深度不足 2，就建在根节点方向能到达的最高点
            int pos = x;
            for (int k = 0; k < 2; k++) {
                if (parent[pos] != 0) {
                    pos = parent[pos];
                }
            }

            if (!hasStation[pos]) {
                hasStation[pos] = 1;
                answer++;

                int p = parent[pos];
                int gp = p ? parent[p] : 0;

                if (p) childStation[p]++;
                if (gp) grandChildStation[gp]++;
            }
        }
    }

    cout << answer << '\n';

    return 0;
}
```
